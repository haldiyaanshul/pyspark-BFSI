### Case 3
we need to modify the content in particular column so we can also use the withColumn with regesp_replace(col('columnname'), "original value", "Modified value or new value which we need to replace") 

```python
df.withColumn('marital' , regexp_replace(col("marital"), 'single', 'Bachlor')) \
.withColumn('martial' , regexp_replace(col("marital"), 'married', 'Married'  )  ) \
.withColumn('martial' , regexp_replace(col("marital"), 'divorced', 'Divorced' )  ) \
.withColumn('martial' , regexp_replace(col("marital"), 'unknown', 'Unknown'   )  ).display()      
              
```

### Sort/orderBy

If we need to organise my data in asc or desc order based on one or more than one columns then we will use the orderBy and Sort

### Case -1 
we need to sort the data based on asc or desc
.sort().asc() .desc()

```python
df.sort(col('age').asc()).display()
```

```python
df.sort(col("age").desc()).display()
```

Sorting based on multiple columns

```python
df.sort(['age', 'duration'], ascending=[True, False]).display()
```

```python
df.select(col('age'), col('job'), col('marital'), col('education'), col("duration")) \
    .sort(["age", "duration"], ascending=[True, False]) \
    .filter(col("marital") == "married") \
    .withColumn("marital_status", regexp_replace(col("marital"), "married", "Fielding Set") )\
    .withColumn('After_Married' , concat(col("marital"), col("marital_status"))) \
    .display()
```

### Limit
If we have alot of data and we only need some few records in that scenario we can use the limit and take only given particular records which is required suppose if we want 100 records only then it will fetch 100 data and so on....

```python
df.limit(10).display()
```

```python
df.select(col('age'), col('job'), col('marital'), col('education'), col("duration")) \
    .sort(["age", "duration"], ascending=[True, False]) \
    .filter(col("marital") == "married") \
    .withColumn("marital_status", regexp_replace(col("marital"), "married", "Fielding Set") )\
    .withColumn('After_Married' , concat(col("marital"), col("marital_status"))) \
        .limit(7) \
    .display()
```

```python
df.select(col('age'), col('job'), col('marital'), col('education'), col("duration")) \
    .sort(["age", "duration"], ascending=[True, False]) \
    .filter((col("marital") == "married") & (col("job") == "admin.") & (col("education") == "basic.9y")) \
    .withColumn("marital_status", regexp_replace(col("marital"), "married", "Fielding Set") )\
    .withColumn('After_Married' , concat(col("marital"), col("marital_status"))) \
        .limit(7) \
    .display()
```
